# Exploratory Data Analysis (EDA) on Shopping Trends Dataset

This notebook provides a comprehensive exploration of a shopping trends dataset. Each step corresponds to specific questions that guide the analysis, covering customer demographics, purchasing behavior, seasonal trends, and more.

**Dataset Overview**:
- Number of Rows: 3900
- Number of Columns: 19
- Features: Customer ID, Age, Gender, Item Purchased, Category, Purchase Amount (USD), Location, Size, Color, Season, Review Rating, Subscription Status, Payment Method, Shipping Type, Discount Applied, Promo Code Used, Previous Purchases, Preferred Payment Method, Frequency of Purchases
- Sample Data: See below for a snapshot.

Let's dive into the analysis!

```python
# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset and display a sample
data = pd.read_csv('shopping_trends.csv')
data.head()

# Display dataset info
data.info()
```

**1. What are the basic characteristics of the data?**

The initial exploration will include data types, missing values, and unique value counts to assess data quality and structure.

```python
data.info()
data.describe(include='all')
```

**2. Are there any missing values in the dataset?**

```python
data.isnull().sum()
```

**5. Which locations have the highest number of purchases?**

```python
location_counts = data['Location'].value_counts().head(10)
sns.barplot(x=location_counts.values, y=location_counts.index)
plt.title('Top 10 Locations by Purchase Count')
plt.xlabel('Number of Purchases')
plt.ylabel('Location')
plt.show()
```

**6. What are the most commonly purchased items and categories?**

```python
top_items = data['Item Purchased'].value_counts().head(10)
sns.barplot(x=top_items.values, y=top_items.index)
plt.title('Top 10 Purchased Items')
plt.xlabel('Frequency')
plt.ylabel('Item Purchased')
plt.show()

top_categories = data['Category'].value_counts()
sns.barplot(x=top_categories.values, y=top_categories.index)
plt.title('Purchase Frequency by Category')
plt.xlabel('Frequency')
plt.ylabel('Category')
plt.show()
```

**7. How do discounts and promo codes affect purchase amounts?**

```python
data['Purchase Amount (USD)'] = data['Purchase Amount (USD)'].replace('[\$,]', '', regex=True).astype(float)
sns.boxplot(x='Discount Applied', y='Purchase Amount (USD)', data=data)
plt.title('Effect of Discounts on Purchase Amount')
plt.xlabel('Discount Applied')
plt.ylabel('Purchase Amount (USD)')
plt.show()

sns.boxplot(x='Promo Code Used', y='Purchase Amount (USD)', data=data)
plt.title('Effect of Promo Codes on Purchase Amount')
plt.xlabel('Promo Code Used')
plt.ylabel('Purchase Amount (USD)')
plt.show()
```

**8. How do purchase patterns vary across seasons?**

```python
season_counts = data['Season'].value_counts()
sns.barplot(x=season_counts.index, y=season_counts.values)
plt.title('Purchases by Season')
plt.xlabel('Season')
plt.ylabel('Frequency')
plt.show()
```

**9. Which shipping types are most commonly chosen?**

```python
shipping_counts = data['Shipping Type'].value_counts()
sns.barplot(x=shipping_counts.values, y=shipping_counts.index)
plt.title('Shipping Preferences')
plt.xlabel('Frequency')
plt.ylabel('Shipping Type')
plt.show()
```

```python

# Enhanced Gender Ratio
gender_counts = data['Gender'].value_counts()
sns.barplot(x=gender_counts.index, y=gender_counts.values, palette='pastel')
plt.title('Gender Distribution of Customers', fontsize=14)
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Count', fontsize=12)
for i, val in enumerate(gender_counts.values):
    plt.text(i, val + 5, f'{val / gender_counts.sum() * 100:.1f}%', ha='center', fontsize=10)
plt.show()
```

```python
# Purchase Amount Analysis
data['Purchase Amount (USD)'] = data['Purchase Amount (USD)'].replace('[\$,]', '', regex=True).astype(float)
category_season = data.groupby(['Category', 'Season'])['Purchase Amount (USD)'].mean().unstack()

plt.figure(figsize=(10, 6))
category_season.plot(kind='bar', figsize=(12, 6), cmap='viridis', edgecolor='black')
plt.title('Average Purchase Amount by Category and Season', fontsize=14)
plt.ylabel('Average Purchase Amount (USD)', fontsize=12)
plt.xlabel('Category', fontsize=12)
plt.legend(title='Season', fontsize=10)
plt.tight_layout()
plt.show()
```

```python
# Correlation Heatmap
numerical_cols = data.select_dtypes(include=['int64', 'float64']).columns
correlation_matrix = data[numerical_cols].corr()

plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap of Numerical Features', fontsize=14)
plt.show()
```
